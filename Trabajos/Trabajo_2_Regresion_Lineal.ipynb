```python
# -*- coding: utf-8 -*-
"""
Created on Sat Dec  5 09:07:08 2020

@author: usuario
"""
import numpy as np
from sklearn import datasets, linear_model
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import math as mt

def datos(): 
    x = np.random.random(20)*10 
    y = np.random.random(20)*10 
    return x.reshape((20,1)), y.reshape((20,1)) 

X=np.array(datos()[0])
Y=np.array(datos()[1])
plt.plot(np.array(datos()[0]),np.array(datos()[1]),'x')
plt.title('Datos')
plt.grid(True)
plt.show()
ex=sum(X)
ey=sum(Y)
exy=sum(X*Y)
exx=sum(X*X)
lon=len(X)
```

```python
m=(lon*exy-ex*ey)/(lon*exx-mt.pow(abs(ex),2))
b=(ey*exx-ex*exy)/(lon*exx-mt.pow(abs(ex),2))
ecua=""
m=round(m[0],4)
b=round(b[0],4)
if (b < 0):
    ecua='y = {}x {}'
else:
    ecua='y = {}x + {}'
print(ecua.format(m,b))
```

```python
fu=lambda x: m*x+b
li=np.arange(min(X)-0.0,max(X)+5.0,0.5)
plt.title('Datos')
plt.plot(X,Y,'x')
plt.plot(li,fu(li))
plt.grid(True)
plt.show()
```

```python
boston = datasets.load_boston()

x = boston.data[:, np.newaxis, 7]
y = boston.target
plt.scatter(x, y)
plt.title('Datos')
plt.grid(True)
plt.show()
```

```python
rl = LinearRegression()
rl.fit(x.reshape(-1,1), y) 
print ('Valor de la ecuaci√≥n')
print('a = ' + str(rl.coef_) + ', b = ' + str(rl.intercept_))
fu=lambda x: rl.coef_*x+rl.intercept_
li=np.arange(min(x),max(x))
plt.scatter(x,y,label='data', color='red')
plt.plot(li,fu(li))
plt.grid(True)
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
rl.fit(x_train, y_train)
y_pred = rl.predict(x_test)

print('Precisi√≥n del modelo:')
print(rl.score(x_train, y_train))
```
