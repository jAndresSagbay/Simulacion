
# Covid-19 en Ecuador

# Formulación del Problema
En estos ulitmos meses hemos visto como la pandemia ha sido propagada en nuestro país, tiene a crecer a diarios, y es por esto que gracias a la simulación, nos planteamos deinir ecuaciones que nos ayuden a enteder el crecicimiento de esta enfermedad, para ello, nos ayudaremos en los dataset existentes.

# Definición del sistema
Se pretende establecer una ecuacion que nos permita simular como se dieron los contagios y muertes en el pais desde el dia 1 que fue el 1 de marzo 2020, fecha desde la cual comenzo la infección en Ecuador

# Formulación del modelo
Para formular el modelo tomamos datos reales de los infectados y muertos por Covid-19 en Ecuador. Se tomaran datos solo de cifras relacionadas a los contagios y las muertes dadas hasta la fecha.Para obtener la ecuación deseado aplicaremos los modelos de Regresión Lineal y Regresión Logarítmica.

# Colección de datos
Los datos para trabajar son mediciones obtenidas en el sistema real de todos los países del mundo, donde se encuentra todas la variables relacionadas con el covid-19.

# Implementación del modelo en el ordenador
Utilizaremos el lenguaje de programación Python con las libería: Pandas, Numpy, Sklearn.

```python
import pandas as pd
import numpy as np
from datetime import datetime,timedelta
from sklearn.metrics import mean_squared_error
from scipy.optimize import curve_fit
from scipy.optimize import fsolve
from sklearn import linear_model
import matplotlib.pyplot as plt

url = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'
df = pd.read_csv(url).fillna(0)
df
```

```python
df= df.loc[(df['location'] == 'Ecuador') & (df['total_cases'] != 0)] # filtrando por pais y no ceros 
df=df[['date','total_cases','total_deaths','location']]
x=np.arange(1,len(df)+1,1) # arreglo de x lo creo para simular el numero del dia y el numero de casos
y=np.array(df.values[:,1])
y1=np.array(df.values[:,2])
FMT = '%Y-%m-%d'
date = df['date']
df['date'] = date.map(lambda x : (datetime.strptime(x, FMT) - datetime.strptime("2020-01-01", FMT)).days)
df
```

```python
df.plot(x ='date', y='total_cases', grid=True, title='Número de Casos')
```

```python

x=np.arange(1,len(df)+1,1) 
y=np.array(df.values[:,1])
y1=np.array(df.values[:,2])

regr = linear_model.LinearRegression()
 
# Entrenamos nuestro modelo
regr.fit(np.array(x).reshape(-1, 1) ,y)

# Veamos los coeficienetes obtenidos, En nuestro caso, serán la Tangente
print('Coefficients: \n', regr.coef_)
# Este es el valor donde corta el eje Y (en X=0)
print('Independent term: \n', regr.intercept_)

y_prediccion = regr.predict([[360]])
```

```python
plt.scatter(x, y)
x_real = np.array(range(50, 350))
plt.title('Número de Casos');
plt.grid(True)
plt.plot(x_real, regr.predict(x_real.reshape(-1, 1)), color='green')
plt.show()
```

```python

plt.figure(figsize=(8, 8))
plt.scatter(x,y,label='Real', color='blue')
plt.title('Número de Casos en 50 dias');
plt.grid(True)
x_real=np.arange(1,len(df)+51,1)
plt.plot(x_real, regr.predict(x_real.reshape(-1, 1)))
plt.show()
```

```python
def modelo_logistico(x,a,b):
    return a+b*np.log(x)
exp_fit = curve_fit(modelo_logistico,x.astype('float'),y.astype('float')) #Extraemos los valores de los paramatros
print(exp_fit)
```

```python

pred_x = list(range(min(x),max(x)+30)) 
plt.rcParams['figure.figsize'] = [7, 7]
plt.rc('font', size=14)
# Real data
plt.scatter(x,y,label="Datos Reales",color="red")
# Predicted exponential curve
plt.plot(pred_x, [modelo_logistico(i,exp_fit[0][0],exp_fit[0][1]) for i in pred_x], label="Modelo Logistico" )
plt.legend()
plt.grid(True)
plt.xlabel("Desde Marzo 2020")
plt.ylabel("Personas infectadas")
plt.ylim((min(y)*0.9,max(y)*3.1))
plt.show()
```
